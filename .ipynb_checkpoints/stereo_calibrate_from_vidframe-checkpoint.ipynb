```python
import os
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
test_vid_folder = 'C:\\Users\\dklev\\Dropbox (University of Michigan)\\MED-LeventhalLab\\Burgess_data\\mouse_SR_videos_to_analyze\\mouse_SR_videos_tocrop\\GFP9\GFP9_202201\GFP9_20220127'
os.chdir(test_vid_folder)
vid_lists = [glob.glob(os.path.join(test_vid_folder,'*_cam01.avi'))]
vid_lists.extend([glob.glob(os.path.join(test_vid_folder,'*_cam02.avi'))])
```

```python
vid_cap_objs = [cv2.VideoCapture(vid_lists[i_list][0]) for i_list in range(2)]
```

```python
# load paired images
img = []
for i_vid, vc in enumerate(vid_cap_objs):
    _, frame = vc.read()
    if i_vid == 0:
        frame = cv2.rotate(frame, cv2.ROTATE_180)
    img.append(frame)
```

```python
# h_fig = plt.figure(figsize = (20,5))
fig, ax = plt.subplots(1,2, figsize=(15,6))
for i_cam in range(2):
    ax[i_cam].imshow(img[i_cam])
    ax[i_cam].set_yticklabels([])
    ax[i_cam].set_xticklabels([])
```
