```python
%matplotlib widget
import pickle
import os
from matplotlib import pyplot as plt
import cv2
import numpy as np
import sys

sr_path = r'C:\Users\dklev\OneDrive\Documents\GitHub\skilled_reaching_pipeline'
# sr_path = r'/home/dleventh/PycharmProjects/skilled_reaching_pipeline'

sys.path.append(sr_path)
import cameras

calib_files_path = r'C:\Users\dklev\Dropbox (University of Michigan)\MED-LeventhalLab\data\dLight\calibration_files'
calib_vids_path = r'C:\Users\dklev\Dropbox (University of Michigan)\MED-LeventhalLab\data\dLight\calibration_videos'

def read_pickle(fname):
    with open(fname, "rb") as handle:
        return pickle.load(handle)

def read_frame(vid_name, frame_num):
    
    cap = cv2.VideoCapture(vid_name)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
    
    ret, img = cap.read()
    
    cap.release()
    
    return img

def get_corners(cal_data, test_frame):
    frame_corners = []
    for i_view in range(3):
        n_rows = len(cal_data['all_rows'][i_view])
        frame_nums = [row['framenum'] for row in cal_data['all_rows'][i_view]]
        
        if test_frame in frame_nums:
            row_idx = frame_nums.index(test_frame)
            
            corners = cal_data['all_rows'][i_view][row_idx]['corners']
        else:
            corners = None
        frame_corners.append(corners)
        
    return frame_corners        
```

```python
cal_file = 'calibrationdata_20230418_12-04-57_b01.pickle'
# GridCalibration_b01_20230503_11-46-08.avi
# vid_name = cal_file.replace('calibrationdata', 'GridCalibration')
name_parts = cal_file.split('_')
vid_name = '_'.join(('GridCalibration', name_parts[3][:3], name_parts[1], name_parts[2] + '.avi'))
calib_month = name_parts[1][:6]
calib_month_folder = 'calibration_files_' + calib_month
vid_month_folder = 'calibration_videos_' + calib_month
calib_file_month_path = os.path.join(calib_files_path, calib_month_folder)
calib_vid_month_path = os.path.join(calib_vids_path, vid_month_folder)

cal_path = os.path.join(calib_file_month_path, cal_file)
cal_vid_path = os.path.join(calib_vid_month_path, vid_name)

cal_data = read_pickle(cal_path)
```

```python
i_frame = 100
img = read_frame(cal_vid_path, i_frame)

fig = plt.figure()
ax = fig.add_subplot()
ax.imshow(img)

frame_corners = get_corners(cal_data, i_frame)

for i_view in range(3):
    if frame_corners[i_view] is not None:
        to_plot = np.squeeze(frame_corners[i_view])
        for i_pt, pt in enumerate(to_plot):
            ax.text(to_plot[i_pt, 0], to_plot[i_pt, 1], '{:d}'.format(i_pt))        
```
